```python
# Dependencies
import pandas as pd
from datetime import datetime
```

```python
# Store filepath in a variable
file_one = "SPY_Exchange_Traded_Fund_OHLC_CandleStickData.csv"
file_two = "sqlify_presidential_tweeets.csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
file_one_df = pd.read_csv(file_one, encoding="ISO-8859-1")
file_one_df.head()
```

```python
file_two_df = pd.read_csv(file_two, encoding="UTF=8")
file_two_df.head()
```

```python
# Using .rename(columns={}) in order to rename columns
renamed_df = file_two_df.rename(columns={"created_at":"Date", "text":"Tweet"})
renamed_df.head()
```

```python
dropped = renamed_df[["Date", "Tweet"]]
dropped.head()
```

```python
# Complete the call to convert the date column
dropped['Date'] =  pd.to_datetime(dropped['Date'], format='%a %Y %m %d')

# Confirm the date column is in datetime format
print(dropped.info())
```

```python
merge_table = pd.merge(file_one_df, dropped, on="Date", how="outer")
merge_table.head()
```

```python
# Export file as a CSV, without the Pandas index, but with the header
dropped.to_csv("just_tweets.csv", index=False, header=True)
```
